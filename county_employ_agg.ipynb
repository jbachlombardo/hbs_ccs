```python
df_geo['State_name'] = df_geo['STABBR'].map(state_code_dict)
```

```python
df_geo_US = df_geo[df_geo['State_name'].notnull()]
```

```python
county_fnames = glob('C:/Users/jbachlombardo/Documents/Community Colleges/Data/2017.annual.by_area/*.csv')
```

```python
county_state_names_match = list()
```

```python
for c, s in zip(df_geo_US['County'], df_geo_US['State_name']) :
    cs_name = c + ', ' + s
    county_state_names_match.append(cs_name)
```

```python
county_fnames_use = list()
```

```python
for c in county_fnames :
    for m in county_state_names_match :
        if m in c :
            county_fnames_use.append(c)
```

```python
county_fnames_use = list(np.unique(county_fnames_use))
```

```python
# unique_NAICS = list(twodig_match['NAICS Code'].unique())
# unique_NAICS = list(foudig_match['NAICS Code'].unique())
unique_NAICS = list(threedig_match['NAICS Code'].dropna().unique())
```

```python
unique_NAICS = [str(int(x)) for x in unique_NAICS]
```

```python
def get_county_totals(data) :
    fips = data['area_fips'].unique().item()
    county_state = data['area_title'].unique().item()
    grouped = data[data['industry_code'].isin(unique_NAICS)].groupby('industry_code', as_index = False)[['annual_avg_emplvl', 'total_annual_wages']].sum()
    grouped['annual_avg_wkly_wage'] = grouped['total_annual_wages'] / grouped['annual_avg_emplvl']
    grouped['FIPS'] = fips
    grouped['County_State'] = county_state
    return grouped
```

```python
all_counties = pd.DataFrame()
```

```python
for f in county_fnames_use :
    single_county = pd.read_csv(f)
    single_county = get_county_totals(single_county)
    all_counties = pd.concat([all_counties, single_county])
```
