```python
six_dig_degs = pd.read_csv('C:/Users/jbachlombardo/Documents/Community Colleges/Data/IPEDS/c2017_a.csv')
```

```python
ccs_sixdig = six_dig_degs[(six_dig_degs['UNITID'].isin(schools_list)) & (six_dig_degs['AWLEVEL'] < 4)]
```

```python
ccs_sixdig['CIPCODE'] = ccs_sixdig['CIPCODE'].apply(lambda x: int(str(x).replace('.', '')))
```

```python
ccs_sixdig_streamlined = ccs_sixdig.groupby(['UNITID', 'CIPCODE'], as_index = False)['CTOTALT'].sum()
```

```python
ccs_sixdig_streamlined = ccs_sixdig_streamlined[ccs_sixdig_streamlined['CIPCODE'] != 990]
```

```python
pivoted_sixdegs = pd.pivot_table(data = ccs_sixdig_streamlined[ccs_sixdig_streamlined['CTOTALT'] != 0], columns = 'CIPCODE', index = 'UNITID', values = 'CTOTALT')
```

```python
pivoted_sixdegs = pivoted_sixdegs.rename_axis(None, axis = 1)
```

```python
totals = pivoted_sixdegs.sum(axis = 1)
```

```python
for c in pivoted_sixdegs.columns :
    newcol = str(c) + '_%'
    pivoted_sixdegs[newcol] = pivoted_sixdegs[c] / totals
```

```python
pivoted_sixdegs.to_csv('C:/Users/jbachlombardo/Documents/Community Colleges/Data/IPEDS/six_deg_ccs_wpercs_no990 051419.csv')
```
